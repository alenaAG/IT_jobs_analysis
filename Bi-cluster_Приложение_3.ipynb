```python
def intersection(lst1, lst2): 
    return list(set(lst1) & set(lst2))

def union(lst1,lst2):
    return list(set(lst1).union(set(lst2)))

def sim(lst1, lst2):
    if ((len(lst1)+len(lst2))/2)!=0:
        return len(intersection(lst1,lst2))/((len(lst1)+len(lst2))/2)
    else: 
        return 0.0
```

```python
def sum_iW(i,W,res):
    r = 0
    for el in W:
        r+=res[i][el]
    return r

def sum_VW(V,W,res):
    r = 0
    for el in V:
        for el2 in W:
            r+=res[el][el2]
    return r

def sum_jV(V,j,res):
    r = 0
    for el in V:
        r+=res[el][j]
    return r
```

```python
import numpy
def del_(r,V,W,row,col):

    # r- data matrix
# V set of rows, W set of columns
# row, col -row/col being added or removed
# val - change of the contribution
# z  - row/column to be added (+1) or removed (-1)
    
    n = len(r)
    m = len(r[0])
    nv=len(V)
    nw=len(W)
    if nw*nv == 0:
        print('Both sets in a box must be non-empty!')
        return val,z
    
    otrow =list(set(range(n))-set(V))
    otcol =list(set(range(m))-set(W))
    rVW = sum_VW(V,W,r)
    if len(col)*len(row) != 0:
        print('Either row or col must be 0 at changing the box')
        return val,z
    
    if len(col) + len(row) == 0:
        print('Either row or col must be not 0 at changing the box')
        return val,z
    
    if len(row) > 0:
        row = row[0]
        sr= sum_iW(row,W,r)
        it= 2*sr*rVW
        if row in V:
            z=- 1
        else:
            z=1

        val=z*it + sr ** 2 - z*rVW**2 / nv
        if nv + z == 0:
            val=- 1
        else:
            val=val / (nw*(nv + z))
    else:
        if len(col) > 0:
            col = col[0]
            sc=sum_jV(V,col,r)
            it=2*sc*rVW
            if col in W:
                z=- 1
            else:
                z=1
            val=z*it + sc ** 2 - z*rVW**2 / nw
            if nw + z == 0:
                val=- 1
            else:
                val=val / (nv*(nw + z))
        else:
            print('No col or row supplied!!!')
            return val,z
    
    return val,z
```

```python
def nbox(r,V,W):
    
    n = len(r)
    m = len(r[0])
    nv=len(V)
    nw=len(W)
    la=sum_VW(V,W,r) / nv*nw
    cont=nv*nw*la ** 2
    row=[]
    cnr=- 1000
    for ii in range(n):
        v,z=del_(r,V,W,[ii],[])
        if v > cnr:
            cnr=v
            zr=z
            row=ii
    col=[]
    cnc=- 1000
    for jj in range(m):
        v,z=del_(r,V,W,[],[jj])
        if v > cnc:
            cnc=v
            zc=z
            col=jj
    if cnr > cnc:
        if cnr > 0:
            if zr > 0:
                Vn=union(V,[row])
            else:
                Vn = list(set(V)-set([row]))
            Wn=W
            dcn=cnr
        else:
            Vn=V
            Wn=W
            dcn=0
    else:
        if cnc > 0:
            if zc > 0:
                Wn=union(W,[col])
            else:
                Wn=list(set(W)-set([col]))
            Vn=V
            dcn=cnc
        else:
            Vn=V
            Wn=W
            dcn=0
    print(Vn,Wn,dcn)
    return Vn,Wn,dcn
```

```python
def brc(r,row,col):


    # r- data matrix
# V set of rows, W set of columns
# la intensity, co contribution
# row, col -row/col being added or removed
# val - change of the contribution
# z  - row/column added (+1) or removed (-1)
    counf=0
    n = len(r)
    m = len(r[0])
    ff=0
    
    if len(col)*len(row) != 0:
        print('Either row or col must be 0 at changing the box')
        return V,W,la,co
    
    if len(col) + len(row) == 0:
        print('Either row or col must be not 0 at changing the box')
        return V,W,la,co
    
    if len(row) > 0:
        V=row
        W = [np.where(r[row][:]>0)[0][0]]
        print(W)
        if len(W)==0:
            ff=1
    else:
        if len(col) > 0:
            W=col
            V=[np.where(r[:][col]>0)[1][0]]
            print(W)
            if len(V)==0:
                ff=1
        else:
            print('No kidding, either row or colomn,not both!')
            return V,W,la,co
    
    if ff > 0:
        V=[]
        W=[]
        la=0
        co=0
    else:
        la=sum_VW(V,W,r) / (len(V)*len(W))
        co=(la ** 2)/len(V)/len(W)
        dd=1
        while (dd > 0):
            V,W,dd=nbox(r,V,W)
            co=co + dd
            counf=counf + 1
            print('count = '+str(counf))
        la=sum_VW(V,W,r) / (len(V),len(W))    
    return V,W,la,co
```
