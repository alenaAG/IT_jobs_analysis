# Сбор всех выделенных списков в один файл для работы

```python
import pandas as pd
data = pd.read_csv('data_forprint.csv',encoding='utf8',index_col=None)
data= data.drop('Unnamed: 0',axis=1)
data.head(5)
```

```python
req = pd.read_csv('data_with_req_refined_2.csv',encoding='utf8',index_col=None)
req= req.drop('Unnamed: 0',axis=1)
req.head(5)
```

```python
resp = pd.read_csv('data_with_resp_refined.csv',encoding='utf8',index_col=None)
resp= resp.drop('Unnamed: 0',axis=1)
resp.head(5)
```

```python
cond = pd.read_csv('data_with_cond_refined.csv',encoding='utf8',index_col=None)
cond= cond.drop('Unnamed: 0',axis=1)
cond.head(5)
```

```python
data['req_refined']=req['requirements_refined']
data['req_idxs']=req['req_idxs']
data['resp_refined']=resp['responsibilities_refined']
data['resp_idxs']=resp['resp_idxs']
data['cond_refined']=cond['conditions_refined']
data['cond_idxs']=cond['cond_idxs']
```

```python
data.to_csv('data_for_clustering.csv')
```
