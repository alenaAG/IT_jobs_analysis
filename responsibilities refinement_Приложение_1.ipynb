# Выделение списка обязанностей из вакансии

```python
import pandas as pd
data = pd.read_csv('data_forprint.csv',encoding='utf8',index_col=None)
data= data.drop('Unnamed: 0',axis=1)
data.head(5)
```

```python
from east.asts import base
import operator
import numpy as np
import re
```

```python
unique_data = data[['ad_id','responsibilities']]
unique_data=unique_data.drop_duplicates()
unique_data.head(5)
```

```python
pat = re.compile(r'[^A-Za-z0-9 \-\n\r.,;!?А-Яа-я]+')
unique_data.responsibilities = [re.sub(pat,' ',x.replace('\n',' ')).lower() for x in unique_data.responsibilities.values[::]]
```

```python
len(unique_data)
```

```python
data_resp = unique_data.responsibilities[::].values
idxs = unique_data.ad_id[::].values
f = open('resp_picked.txt', 'r')
resp_picked = f.readlines()
f.close()
data_resp= list(data_resp)
```

```python
data_resp[0:4]
```

```python
ast = base.AST.get_ast(data_resp)
```

```python
resp_picked = list(map(lambda x: re.sub(pat,' ',x.replace('\n',' ')).lower(), resp_picked))
```

```python
resp_ast = {}
for el in resp_picked:
    try:
        resp_ast[el] = ast.score(el)
    except:
        print('oops')
        print(el)
```

```python
resp_ast = reversed(sorted(resp_ast.items(), key=operator.itemgetter(1)))
```

```python
for el in resp_ast:
    print(el)
```

```python
def make_substrings(tokens, k=4):
    tokens = tokens.split()
    for i in range(max(len(tokens) - k + 1, 1)):
        yield ' '.join(tokens[i:i + k])
        
def get_relevance_matrix(texts, strings):

    matrix = np.empty((0, len(strings)), float)

    for text in texts:
        ast = base.AST.get_ast(list(make_substrings(text)))
        row = np.array([ast.score(s) for s in strings])
        matrix = np.append(matrix, [row], axis=0)

    return matrix
```

```python
relevance_matrix = get_relevance_matrix(data_resp, resp_picked)


print("Relevance mairix:")
```

```python
np.savetxt('relevance_matrix_resp.txt',relevance_matrix)
```

```python
mask = relevance_matrix>=0.3
mask
```

```python
idx_x_responsibilities = pd.DataFrame(columns=['ad_id','responsibilities_refined','resp_idxs'])
```

```python
for i in range(len(mask)):
    idx_x_responsibilities = idx_x_responsibilities.append({'ad_id': idxs[i],'responsibilities_refined':np.array(resp_picked)[mask[i]],'resp_idxs':np.where(relevance_matrix[i]>=0.3)}, ignore_index=True)
    print(np.array(resp_picked)[mask[i]])
```

```python
final = pd.merge(data, idx_x_responsibilities, left_on='ad_id', right_on='ad_id', how='left')
```
