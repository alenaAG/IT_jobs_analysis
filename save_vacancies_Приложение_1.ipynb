## Получение вакансий через api

```python
import requests
import json

def get_vacancy_idxs(pageN,idxs):
    url = "https://api.hh.ru/vacancies"
    querystring = {"specialization":"1.395","1.400":"","1.420":"","1.474":"","1.475":"","1.536":"","1.3":"","1.9":"","1.10":"","1.25":"","1.82":"","1.110":"","1.117":"","1.203":"","1.211":"","1.221":"","1.270":"","1.272":"","1.273":"","1.296":"","1.327":"","per_page":"100","page":pageN}
    headers = {
        'User-Agent': "PostmanRuntime/7.15.0",
        'Accept': "*/*",
        'Cache-Control': "no-cache",
        'Postman-Token': "e9113f34-84fc-4028-aa10-08001de3769a,7bc059ec-8fcc-4ef8-8728-70f69f7c68af",
        'Host': "api.hh.ru",
        'accept-encoding': "gzip, deflate",
        'Connection': "keep-alive",
        'cache-control': "no-cache"
        }
    
    response = requests.request("GET", url, headers=headers, params=querystring)
    responses = json.loads(response.text)
    for el in responses['items']:
        idxs.append(el['id'])
        
```

```python
idxs = []
for i in range(0,20):
    get_vacancy_idxs(i,idxs)
```

```python
print(len(idxs))
```

```python
import re
def remove_tags(text):
    return re.sub(r'\<.*\>', '', text)
```

## Парсинг текстов вакансий для выделения списков и удаления html разметки

```python
import pandas as pd
data=[]
data = pd.DataFrame(columns=['domain_id','domain_name','profearea_id','profearea_name','ad_id', 'name', 'description','req','resp','cond','date'])

for idx in idxs:
    url = "https://api.hh.ru/vacancies/"+str(idx)
    print(url)
    headers = {
    'User-Agent': "PostmanRuntime/7.15.0",
    'Accept': "*/*",
    'Cache-Control': "no-cache",
    'Postman-Token': "c16c7cdf-120d-49f7-ba2d-a06d78a3f8f2,384c178b-a538-4c54-ae90-ab4817e9486e",
    'Host': "api.hh.ru",
    'accept-encoding': "gzip, deflate",
    'Connection': "keep-alive",
    'cache-control': "no-cache"
    }
    response = requests.request("GET", url, headers=headers)
    response = json.loads(response.text)
    desc = response['description']
    date = response['published_at']
    
    try:
        spec = response['specializations']
        profarea_id = []
        profarea_name = []
        for el in spec:
            if el['profarea_id']=='1':
                profarea_id.append(el['id'])
                profarea_name.append(el['name'])

        desc_arr = desc.split('<ul>')
        if (len(desc_arr)==4):
            resp = desc_arr[1].split('</ul>')[0].replace('</li>','').split('<li>')[1:]
            for el in resp:
                el = remove_tags(el)
            req = desc_arr[2].split('</ul>')[0].replace('</li>','').split('<li>')[1:]
            for el in req:
                el = remove_tags(el)
            cond = desc_arr[3].split('</ul>')[0].replace('</li>','').split('<li>')[1:]
            for el in cond:
                el = remove_tags(el)
            for i in range(len(profarea_id)):
                data = data.append({'domain_id':'1','domain_name':'Информационные технологии, интернет, телеком','profearea_id':profarea_id[i],'profearea_name':profarea_name[i],'ad_id':idx,'name':response['name'],'description':desc,'req':req,'resp':resp,'cond':cond,'date':date},ignore_index=True)
        elif (len(desc_arr)==3):
            resp = desc_arr[1].split('</ul>')[0].replace('</li>','').split('<li>')[1:]
            for el in resp:
                el = remove_tags(el)
            req = desc_arr[2].split('</ul>')[0].replace('</li>','').split('<li>')[1:]
            for el in req:
                el = remove_tags(el)
            cond = None
            for i in range(len(profarea_id)):
                data = data.append({'domain_id':'1','domain_name':'Информационные технологии, интернет, телеком','profearea_id':profarea_id[i],'profearea_name':profarea_name[i],'ad_id':idx,'name':response['name'],'description':desc,'req':req,'resp':resp,'cond':cond,'date':date},ignore_index=True)
    except:
        print('oops')
```

```python
from datetime import datetime
def to_date(date):
    return datetime.strptime(date.split('T')[0], '%Y-%m-%d')
date = data['date'].apply(to_date)
date = sorted(date)
first_date = date[0]
last_date = date[-1]
```

```python
data_fp['requirements']=data_fp['requirements'].apply('\n'.join)
data_fp['responsibilities']=data_fp['responsibilities'].apply('\n'.join)
data_fp['conditions']=data_fp['conditions'].apply('\n'.join)
```

```python
def elim_com(x):
    try:
        x=x.replace('<p>','')
        x=x.replace('</p>','')
        x=x.replace('<em>','')
        x=x.replace('</em>','')
        return x
    except ValueError:
        return x
```

```python
data_fp['conditions']=data_fp['conditions'].apply(elim_com)
data_fp['requirements']=data_fp['requirements'].apply(elim_com)
data_fp['responsibilities']=data_fp['responsibilities'].apply(elim_com)
```
